```
# Instale a biblioteca necessária (rode esta célula primeiro)
!pip install transformers torch

# --- Célula 2: Importação e Configuração ---
from transformers import pipeline
import time

# Definindo o prompt (entrada) para os modelos
prompt = "O futuro da inteligência artificial no Brasil é"

print(f"Prompt de entrada: '{prompt}'\n")
print("-" * 50)

# --- Modelo 1: GPT-2 ---
print("Carregando o Modelo 1: gpt2...")
start_time = time.time()
generator_1 = pipeline('text-generation', model='gpt2')
resultado_1 = generator_1(prompt, max_length=50, num_return_sequences=1, truncation=True)
tempo_1 = time.time() - start_time

print(f"Resultado GPT-2 ({tempo_1:.2f} segundos):")
print(resultado_1[0]['generated_text'])
print("-" * 50)

# --- Modelo 2: GPT-Neo 125M ---
print("Carregando o Modelo 2: EleutherAI/gpt-neo-125m...")
start_time = time.time()
generator_2 = pipeline('text-generation', model='EleutherAI/gpt-neo-125m')
resultado_2 = generator_2(prompt, max_length=50, num_return_sequences=1, truncation=True)
tempo_2 = time.time() - start_time

print(f"Resultado GPT-Neo ({tempo_2:.2f} segundos):")
print(resultado_2[0]['generated_text'])
print("-" * 50)
```
